```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
from rpy2.robjects import r

# Import necessary R packages
grdevices = importr('grDevices')
utils = importr('utils')

# Check if the C50 package is installed, if not, install it
try:
    c50 = importr('C50')  # Import the C50 package
except Exception as e:
    print("C50 package not found. Installing...")
    utils.install_packages('C50')
    c50 = importr('C50')  # Try importing again

# Load the iris dataset from R
iris = r['iris']

# Prepare input and output for the model
input_data = iris.rx(True, ro.IntVector(range(1, 5)))  # Select columns 1 to 4 using R-style indexing

# Convert output_data to a factor
output_data = r.factor(iris.rx(True, 5))  # Select the fifth column as output (Species)

# Check dimensions of input and output data
print(f"Input data dimensions: {input_data.nrow} rows, {input_data.ncol} columns")
print(f"Output data dimensions: {len(output_data)} rows")

# Build the C5.0 decision tree model (Pruned)
try:
    model2 = c50.C5_0(input_data, output_data, control=c50.C5_0Control(noGlobalPruning=False))
    r.assign("model2", model2)  # Assign the model to the R environment
    print("Model created and assigned successfully.")
except Exception as e:
    print(f"Error in model creation: {e}")
    model2 = None

# Proceed only if the model was created successfully
if model2:
    # Open a PNG device to save the decision tree plot
    grdevices.png(file="iris_decision_tree_pruned.png", width=800, height=600)

    # Plot the decision tree
    r('plot(model2, main="C5.0 Decision Tree - Pruned")')

    # Close the PNG device to save the plot
    grdevices.dev_off()

    # Output message
    print("C5.0 pruned decision tree saved as 'iris_decision_tree_pruned.png'")
else:
    print("Model creation failed. Plotting skipped.")
```

```python
import rpy2.robjects as ro
import pandas as pd
import numpy as np
from rpy2.robjects import pandas2ri

# Activate the pandas-to-R DataFrame conversion
pandas2ri.activate()

# Load the iris dataset as a pandas DataFrame
iris = ro.conversion.rpy2py(ro.r['iris'])

# Prepare input data for the model
input_data = iris.iloc[:, :4]  # First four columns (features)

# Convert the input DataFrame to a NumPy array
input_data_np = input_data.values

# Create R vectors for each column and then create the R DataFrame
input_data_r = ro.r['data.frame'](
    Sepal_Width=ro.FloatVector(input_data_np[:, 1]),
    Petal_Length=ro.FloatVector(input_data_np[:, 2]),
    Petal_Width=ro.FloatVector(input_data_np[:, 3])
)


# Ensure the model2 object is correctly assigned in the R environment
try:
    model2 = ro.r['model2']
except Exception as e:
    print(f"Error accessing model2: {e}")
    model2 = None

if model2:
    # Perform predictions for the entire iris dataset using model2
    try:
        predicted = ro.r['predict']

        # Convert the R factor of predicted values to a pandas Series
        predicted_py = pd.Series(ro.conversion.rpy2py(predicted), name='predictedC501')

        # Add the predicted values as a new column to the pandas DataFrame
        iris['predictedC501'] = predicted_py

        # Ensure the 'Species' and 'predictedC501' columns are of the same type
        iris['Species'] = iris['Species'].astype(str)
        iris['predictedC501'] = iris['predictedC501'].astype(str)

        # Filter rows where actual species doesn't match predicted species
        mismatch = iris[iris['Species'] != iris['predictedC501']]

        # Display the mismatched rows
        print(mismatch)
    except Exception as e:
        print(f"Error during prediction: {e}")
else:
    print("Model creation failed or model2 not found. Predictions skipped.")
```

**This is an example of a notebook to demonstrate concepts of Data Science. In this example we will do some exploratory data analysis on the famous Iris dataset.

The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). These measures were used to create a linear discriminant model to classify the species. The dataset is often used in data mining, classification and clustering examples and to test algorithms.**
